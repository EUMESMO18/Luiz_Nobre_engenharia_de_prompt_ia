### Verificador de Temperatura

Este código solicita a temperatura ao usuário e, se ela exceder um limite predefinido (neste caso, 30 graus Celsius), exibe um alerta.

```python
# Solicita ao usuário que digite a temperatura atual em Celsius.
temperatura_str = input('Digite a temperatura atual em Celsius: ')

try:
    # Tenta converter a entrada do usuário (que é uma string) para um número decimal (float).
    temperatura = float(temperatura_str)
except ValueError:
    # Se a conversão falhar (por exemplo, se o usuário digitar texto em vez de um número), imprime uma mensagem de erro.
    print('Entrada inválida. Por favor, digite um número.')
    # Define 'temperatura' como None para indicar que a entrada foi inválida e não há temperatura válida para processar.
    temperatura = None

# Verifica se a temperatura é um valor válido (ou seja, não é None).
if temperatura is not None:
    # Se a temperatura for maior que 30 graus Celsius, exibe um alerta de temperatura muito alta.
    if temperatura > 30:
        print(f'Alerta! Temperatura muito alta: {temperatura}°C. Tome cuidado!')
    # Caso contrário, se a temperatura for maior que 25 graus Celsius, exibe que a temperatura está agradável.
    elif temperatura > 25:
        print(f'A temperatura de {temperatura}°C está agradável.')
    # Se nenhuma das condições anteriores for verdadeira (temperatura <= 25 graus Celsius), exibe que a temperatura está baixa.
    else:
        print(f'A temperatura de {temperatura}°C está baixa.')
```
