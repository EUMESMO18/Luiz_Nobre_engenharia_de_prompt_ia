## Missão 1: Automação de Arquivos
### Nome: Luiz Fernando Nobre
Este programa em Python irá escanear uma pasta de origem, identificar a extensão de cada arquivo e movê-los para subpastas correspondentes em um diretório de destino. Por exemplo, todos os arquivos `.pdf` irão para uma pasta chamada `PDF`.

```python
import os
import shutil

def organize_files(source_dir, destination_dir):
    """
    Organiza arquivos de um diretório de origem para um diretório de destino
    baseando-se na extensão do arquivo.

    Args:
        source_dir (str): O caminho para o diretório de origem dos arquivos.
        destination_dir (str): O caminho para o diretório onde os arquivos
                                organizados serão movidos.
    """
    # Verifica se o diretório de origem existe
    if not os.path.exists(source_dir):
        print(f"Erro: O diretório de origem '{source_dir}' não existe.")
        return

    # Cria o diretório de destino se não existir
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)
        print(f"Diretório de destino '{destination_dir}' criado.")

    print(f"\nIniciando organização de arquivos de '{source_dir}' para '{destination_dir}'...")

    # Itera sobre todos os arquivos na pasta de origem
    for filename in os.listdir(source_dir):
        file_path = os.path.join(source_dir, filename)

        # Verifica se é um arquivo (ignora subdiretórios)
        if os.path.isfile(file_path):
            # Extrai a extensão do arquivo e converte para minúsculas
            file_extension = os.path.splitext(filename)[1].lower()

            if file_extension:
                # Remove o ponto da extensão e a converte para maiúsculas para usar como nome da pasta
                folder_name = file_extension[1:].upper() if file_extension.startswith('.') else file_extension.upper()
                target_folder = os.path.join(destination_dir, folder_name)

                # Cria a pasta de destino específica para a extensão, se não existir
                if not os.path.exists(target_folder):
                    os.makedirs(target_folder)
                    print(f"  - Criada pasta: '{target_folder}'.")

                destination_path = os.path.join(target_folder, filename)

                # Tenta mover o arquivo
                try:
                    shutil.move(file_path, destination_path)
                    print(f"  - Movido '{filename}' para '{target_folder}'.")
                except Exception as e:
                    print(f"Erro ao mover '{filename}': {e}")
            else:
                print(f"  - '{filename}' não possui extensão, será ignorado.")

    print("Organização de arquivos concluída!")

# --- Exemplo de Uso ---

# Caminho da pasta de origem (onde estão os arquivos bagunçados)
source_directory = '/content/minha_pasta_baguncada'

# Caminho da pasta de destino (onde os arquivos serão organizados)
destination_directory = '/content/pasta_organizada'

print("\n--- Preparando ambiente de teste ---")
# Limpa e recria os diretórios para garantir um teste limpo a cada execução
if os.path.exists(source_directory):
    shutil.rmtree(source_directory)
    print(f"Diretório de origem '{source_directory}' limpo.")
if os.path.exists(destination_directory):
    shutil.rmtree(destination_directory)
    print(f"Diretório de destino '{destination_directory}' limpo.")

os.makedirs(source_directory)
print(f"Diretório de origem '{source_directory}' criado.")

# Criando arquivos de teste com diferentes extensões
with open(os.path.join(source_directory, 'documento1.pdf'), 'w') as f: f.write('conteudo pdf')
with open(os.path.join(source_directory, 'imagem1.jpg'), 'w') as f: f.write('conteudo jpg')
with open(os.path.join(source_directory, 'relatorio.docx'), 'w') as f: f.write('conteudo docx')
with open(os.path.join(source_directory, 'foto_ferias.png'), 'w') as f: f.write('conteudo png')
with open(os.path.join(source_directory, 'musica.mp3'), 'w') as f: f.write('conteudo mp3')
with open(os.path.join(source_directory, 'sem_extensao'), 'w') as f: f.write('arquivo sem extensao')
print("Arquivos de teste criados no diretório de origem.")

# --- Chamada da função principal para organizar os arquivos ---
# Esta linha executa a organização dos arquivos do diretório de origem para o de destino.
organize_files(source_directory, destination_directory)

print("\nScript de organização de arquivos finalizado.")
```
