### Missão 3: Sistema Autônomo Inicial de Alerta

###Nome: Luiz Fernando Nobre

Esta primeira versão configura um sistema básico para monitorar uma métrica simulada e disparar um alerta no console caso um limiar crítico seja atingido. Inclui funções para verificar a condição e enviar alertas, juntamente com um loop principal para executar o sistema de forma autônoma.

```python
import time # Importa o módulo time para usar a função sleep, que pausa a execução do programa por um número específico de segundos.
import random # Importa o módulo random para simular valores aleatórios

# --- Configuração --- #
CRITICAL_THRESHOLD = 90  # Exemplo de limite para uma métrica como uso de CPU
MONITOR_INTERVAL = 2     # Frequência de verificação da condição (em segundos)

# --- Função para simular a verificação de uma condição --- #
def check_system_condition():
    """
    Simula a verificação de uma métrica do sistema (ex: uso de CPU).
    Retorna True se a condição for crítica, False caso contrário.
    """
    # Em um cenário real, isso leria dados de sensores reais ou métricas do sistema
    current_metric_value = random.randint(0, 100) # Simula uma métrica de 0 a 100
    print(f"Monitorando: O valor atual da métrica é {current_metric_value}")

    if current_metric_value > CRITICAL_THRESHOLD:
        return True, current_metric_value
    else:
        return False, current_metric_value

# --- Função para enviar um alerta via console --- #
def send_console_alert(metric_value):
    """
    Envia uma mensagem de alerta crítica para o console.
    """
    alert_message = f"ALERTA CRÍTICO: O valor da métrica {metric_value} excedeu o limite de {CRITICAL_THRESHOLD}!"
    print("\n" + "="*len(alert_message)) # Imprime uma linha de separação para destacar o alerta
    print(alert_message) # Imprime a mensagem de alerta
    print("="*len(alert_message) + "\n") # Imprime outra linha de separação

# --- Loop autônomo principal --- #
def autonomous_monitor_loop():
    """
    O loop principal que monitora continuamente o sistema e dispara alertas.
    """
    print("Iniciando o sistema de monitoramento autônomo...") # Informa que o monitoramento está começando
    try:
        while True: # Loop infinito para monitoramento contínuo
            is_critical, value = check_system_condition() # Chama a função para verificar a condição do sistema
            if is_critical: # Verifica se a condição retornada é crítica
                send_console_alert(value) # Se for crítica, envia o alerta
            time.sleep(MONITOR_INTERVAL) # Pausa pelo tempo definido em MONITOR_INTERVAL antes da próxima verificação
    except KeyboardInterrupt: # Captura a exceção quando o usuário interrompe o programa (ex: Ctrl+C)
        print("Monitoramento interrompido pelo usuário.") # Informa que o monitoramento foi interrompido
    print("Sistema de monitoramento autônomo encerrado.") # Informa que o sistema foi encerrado

# --- Executa o sistema de monitoramento --- #
autonomous_monitor_loop() # Chama a função principal para iniciar o monitoramento
```
