# CalculadoraSimples

```python
# Imprime uma mensagem de boas-vindas ao usuário
print("Calculadora Simples")

# Solicita ao usuário o primeiro número e o converte para float
num1 = float(input("Digite o primeiro número: "))
# Solicita ao usuário o segundo número e o converte para float
num2 = float(input("Digite o segundo número: "))

# Solicita ao usuário a operação desejada (soma ou subtracao) e converte para minúsculas
operação = input("Escolha a operação (soma ou subtracao): ").lower()

# Inicializa as variáveis resultado e mensagem_operacao
resultado = 0
mensagem_operacao = ""

# Realiza a operação escolhida com base na entrada do usuário
if operação == "soma":
    # Se a operação for 'soma', calcula a soma dos dois números
    resultado = num1 + num2
    # Define a mensagem da operação como 'soma'
    mensagem_operacao = "soma"
elif operação == "subtracao":
    # Se a operação for 'subtracao', calcula a diferença entre os dois números
    resultado = num1 - num2
    # Define a mensagem da operação como 'subtração'
    mensagem_operacao = "subtração"
else:
    # Se a operação for inválida, imprime uma mensagem de erro
    print("Operação inválida. Por favor, escolha 'soma' ou 'subtracao'.")

# Exibe o resultado se uma operação válida foi realizada
if mensagem_operacao:
    # Imprime o resultado formatado, indicando qual operação foi feita e o valor obtido
    print(f"O resultado da {mensagem_operacao} é: {resultado}")
```
